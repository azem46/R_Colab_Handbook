# **Matrisler**
Matrisler, vektörler gibi tek bir tür veri içerir. Buna karşın, vektörler tek boyutlu iken matrisler iki boyutlu veri yapılarıdır. Matris oluşturmanın birden fazla yolu vardır. Eşit uzunluktaki vektörler birleştirilerek matris oluşturulabilir. Sütunların birleştirilmesi için cbind(), satırların birleştirilmesi için rbind() fonksiyonları kullanılır.

```r
#Birden fazla vektör kullanarak matris oluşturmak için
# öğrencilerin boyları için bir vektör, kiloları için başka bir vektör oluşturma
ogrenci_boy <- c(162, 168, 175, 182, 140)
ogrenci_kilo <- c(48, 50, 52, 55, 48)

mm_1 <- cbind(ogrenci_boy, ogrenci_kilo)
mm_1
```

```r
mm_2 <- rbind(ogrenci_boy, ogrenci_kilo)
mm_2
```

```r
class(mm_1) ; class(mm_2)
```

```r
#Tek vektörden istenilen sütün ve satır kadar matris üretmek için 2 sütun 4 satırlık matris
a <-  matrix(c(2,3,5,8,9,45,25,13),nrow=4,ncol=2)
a
```

```r
#Matrisin kaç satır kaç sütundan oluştuğunu öğrenmek için dim() fonksiyonu kullanılır.
dim(a)
```

```r
#matris sıralanmasını yukardan aşağıya değilse soldan sağa göre yapmak için byrow=TRUE ayarını ekliyorum varsayılan FALSE dir.
# İlk başta ilk sütun oluşur sonra sırası ile sütunları oluşturarak gider. Varsayılan satırları oluşturarak gider.
yeni_matris <- matrix(1:20, nrow = 4, ncol = 5, byrow = T)
```

```r
#oluşturulan matrise başlıklar atamak için
rownames(yeni_matris) <- c("Satir1","Satir2","Satir3","Satir4")
colnames(yeni_matris) <- c("sutun1","sutun2","sutun3","sutun4","sutun5")

yeni_matris
```

```r
# sütun ve satırlara isim verme (paste() fonksiyonundan yararlanarak)
colnames(yeni_matris) <- paste("A", c(1:5), sep = "")
yeni_matris
```

```r
rownames(yeni_matris) <- paste("", c(1:4), sep = "")
yeni_matris
```

```r
#Matrisin istenilen elemanına erişmek için 2. satır 3. sütun da bulunan değer nedir?
yeni_matris[2,3]
```

```r
# sadece istenilen satır veye sütundaki tüm değerleri görmek için
yeni_matris[2,]
```

```r
yeni_matris[,5]
```

```r
#1. Satırdan 4. satıra kadar olan satırların son değerlerini verir
yeni_matris[c(1,4),-1]
```

```r
yeni_matris[-1, ] #1. satır harici diğerlerini getirir.
```

```r
yeni_matris[-c(3), ]#3. satır harici diğerlerini getirir.
```

```r
d_matris <- matrix(1:4,nrow = 2,ncol = 2,byrow = TRUE)
d_matris
```

```r
#Transpose işlemi yani satırları sütunlara dönüştürmek için t()fonksiyonu kullanılır.
t(d_matris)
```

```r
#Matrislerde aritmetik işlemler yapılabilir. 5* matris ile matrisin tüm değerleri 5 ile çarpılır.

bes_matris=5*d_matris
bes_matris
```

```r
#Eşit sütun ve satır sayısına sahip matrislerde de aritmetik işlemler yapılabilir.
3*d_matris+d_matris
```

```r
# matris ile devriğinin çarpılması
d_matris %*% t(d_matris)
```

```r
#Matrislerin benzerliğini karşılaştırmak için biribirinin aynısı mı sorgusu yapar benzerse true döner değilse false
identical(d_matris,bes_matris)
```

```r
#iki farklı vektörün seçilen hücre değerlerini karşılaştırdık.
identical(yeni_matris[2,3], yeni_matris[3,2])
```

```r
# m1 matrisinin oluşturulması
matris1 <- matrix(1:15, 3, 5)
matris1

```

```r
# sütun ve satır ortalamaları
colMeans(matris1)
rowMeans(matris1)
```

```r
#Ortalama Hesaplamanın Bir Başka Yöntemi
matris1
mean(matris1[,2])
mean(matris1[3,])
```

```r
mat1 <- matrix(1:15, ncol=3)
mat1
```

```r
#Vektörleride hesaplamayı apply fonksiyonu ile yapma
apply(mat1,MARGIN = 1,FUN = sum)# satır için margine 1 yazıyoruz
apply(mat1,MARGIN = 2,FUN = sum)# sütun için margine 2 yazıyoruz.
```

```r
apply(mat1,MARGIN = 1,FUN = mean)# satır için margine 1 yazıyoruz
apply(mat1,MARGIN = 2,FUN = mean)# sütun için margine 2 yazıyoruz.
```

```r
#sıralama yapmak istersek bir nevi transpose gibi düşünebiliriz.
apply(mat1,MARGIN = 1,FUN = sort)# satır için margine 1 yazıyoruz

```

```r
#Veriyi standartlaştırmak için scale() fonksiyonu kullanılıri Elemanlardan sütun ortalamaları çıkarılarak hesaplanır.
scale(mat1)
```
