# ** Listeler **

Listeler vektörlere benzer ancak vektörlerden farklı olarak herhangi bir veri türünde veri içerebilir. Esnek bir veri yapısıdır. Listelerde hem farklı türden, hem de farklı uzunlukta veriler yer alabilir. Listeler list() fonksiyonu ile oluşturulabilir.Vektöri matris ve diziyi bir arada tutabilir.

```r
# bir listeye koymak için üç ayrı vektör oluşturma
ad <- c("Ad1", "Ad2", "Ad3", "Ad4", "Ad5", "Ad6")
boy <- c(180, 181, 188, 170, 172, 175, 181)
kilo <- c(80, 85, 55)

# üç vektörden oluşan bir liste oluşturma
liste <- list(ad, boy, kilo)
liste
```

```r
#Listenin 1. elemanına ulaşmak için
liste[1]
```

```r
#Tek köşeli parantezde alt bileşenleri döndürülür iç içe köşeli pazantez ile ilgili alt bileşenin değerleri döndürülür.
liste[[1]]
```

```r
#Alt listenin istenilen elemanını görüntülemek için 2. bileşenin 3. elemanını verir 188 i verir

liste[[2]][3]
```

```r
# liste'nin yapısının incelenmesi
str(liste)
```
